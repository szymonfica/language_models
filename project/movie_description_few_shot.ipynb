```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()
```

```python
def generate_description_few_shot(title: str, genres: str, director: str, max_length: int = 250) -> str:
    prompt = f"""Example 1:
    Title: Inception
    Genres: Action, Sci-Fi, Thriller
    Director: Christopher Nolan
    Description: Inception is a complex thriller that takes the viewer on a journey through the world of dreams and reality, where every detail matters.

    Example 2:
    Title: La La Land
    Genres: Musical, Romance, Drama
    Director: Damien Chazelle
    Description: La La Land is a colorful story about dreams, love, and music that touches the hearts of viewers.

    Now:
    Title: {title}
    Genres: {genres}
    Director: {director}
    Description:"""

    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    with torch.no_grad():
        output_ids = model.generate(
            input_ids,
            max_length=len(input_ids[0]) + max_length,
            temperature=0.8,
            do_sample=True,
            top_p=0.95,
            pad_token_id=tokenizer.eos_token_id
        )

    generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    if "Description:" in generated_text:
        generated_description = generated_text.split("Description:")[-1].strip()
    else:
        generated_description = generated_text.strip()

    return generated_description
```

```python
title = "Matrix"
genres = "Action, Sci-Fi"
director = "Lana Wachowski, Lilly Wachowski"
description = generate_description_few_shot(title, genres, director)
print("Generated description (few-shot):")
print(description)
```
